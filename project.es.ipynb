```python
# Código
calificaciones = [8.5,9.2,7.8,8.9,10]
def promedio_calificaciones(calificaciones):
    return sum(calificaciones) / len(calificaciones)
```

#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# TODO
precio_maximo = ds["price"].max()
casa_mas_cara = ds.loc[ds["price"] == precio_maximo]
direccion = casa_mas_cara.iloc[0]["address"]
precio = casa_mas_cara.iloc[0]["price"]
print(f"La casa con dirección {direccion} es la más cara y su precio es de {precio}")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# TODO
precio_minimo = ds["price"].min()
casa_mas_barata = ds.loc[ds["price"] == precio_minimo]
direccion = casa_mas_barata.iloc[0]["address"]
precio = casa_mas_barata.iloc[0]["price"]
print(f"La casa con dirección {direccion} es la más barata y su precio es de {precio}")
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

# Trazar el histograma de precios
ds_arroyo = ds[ds["level5"] == "Arroyomolinos (Madrid)"].copy()
ds_arroyo["price"] = ds_arroyo["price"].astype(float)
prices = ds_arroyo["price"].dropna()
plt.figure(figsize=(8, 5))
plt.hist(prices, bins=20)
plt.title("Histograma de precios - Arroyomolinos (Madrid)")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

Análisis del histograma de precios en Arroyomolinos (Madrid)

El histograma muestra que la mayor parte de los inmuebles se concentra en un rango de precios aproximado entre 200.000 € y 350.000 €, lo que indica que este es el tramo de precios más habitual en la población.

La distribución presenta un sesgo a la derecha, ya que existen algunos inmuebles con precios considerablemente más altos (por encima de los 400.000 €), que generan una cola hacia los valores elevados. Estos casos corresponden probablemente a viviendas unifamiliares o de mayor tamaño.

Asimismo, se observa algún valor muy bajo o atípico, que podría deberse a errores de registro o a inmuebles con características especiales, y que convendría revisar en un análisis más profundo.

En conjunto, el histograma refleja un mercado relativamente homogéneo, dominado por precios medios, con pocos valores extremos.
